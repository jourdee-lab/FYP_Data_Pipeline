```python
import pandas as pd

sas04 = pd.read_csv("https://raw.githubusercontent.com/jourdee-lab/manchester-spatial-analysis-data-lab/refs/heads/main/data/raw/census_1981/sas04_country_of_birth/sas04_1981_combined.csv")
sas07 = pd.read_csv("https://raw.githubusercontent.com/jourdee-lab/manchester-spatial-analysis-data-lab/refs/heads/main/data/raw/census_1981/sas07_employment/sas07_1981_combined.csv")
sas10 = pd.read_csv("https://raw.githubusercontent.com/jourdee-lab/manchester-spatial-analysis-data-lab/refs/heads/main/data/raw/census_1981/sas10_housing/sas10_1981_combined.csv")
sas02 = pd.read_csv("https://raw.githubusercontent.com/jourdee-lab/manchester-spatial-analysis-data-lab/refs/heads/main/data/raw/census_1981/sas02_total_population/sas02_1981_combined.csv")

# Drop zoneid from all but the first so it doesn't duplicate
profile = pd.concat(
    [
        sas04,
        sas07.drop(columns="zoneid"),
        sas10.drop(columns="zoneid"),
        sas02.drop(columns="zoneid"),
    ],
    axis=1
)

profile.to_csv("final_merged_1981.csv", index=False)
print("Manchester 1981 profile saved – all SAS tables merged")
```
